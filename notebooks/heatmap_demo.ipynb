```python
import sys
sys.path.append('..')
import pandas as pd
from ift6758.visualizations import heatmap
```

```python
df = pd.read_csv("../ift6758/data/preprocessed_data/preprocessed_game_data.csv")
df['is_goal'] = df['type'] == 'Goal'
```

```python
df_2020 = df
x_coords = df_2020['xCoord'].to_list()
y_coords = df_2020['yCoord'].to_list()
x_coords_int = []
y_coords_int = []
for x, y in zip(x_coords, y_coords):
    try:
        x = int(x)
        y = int(y)
        x_coords_int.append(x)
        y_coords_int.append(y)
    except:
        pass
print(len(x_coords_int), len(y_coords_int))
heatmap.plot_coords_on_heatmap(x_coords_int, y_coords_int)
```
