```python
import sys
sys.path.append('..')
import pandas as pd
```

```python
df = pd.read_csv('../ift6758/data/preprocessed_data/preprocessed_game_data.csv')
df
```

```python
# making the assumption that a game is 60min

filter_df_shot = df[df['type'].isin(['ShotOnGoal', 'Goal'])]
shot_per_games = filter_df_shot.groupby('id').count()
shot_per_games
average_shots_per_game = shot_per_games['type'].mean()
average_shots_per_game

```

```python
away_team_shot = filter_df_shot.groupby('awayTeam_abbrev')['type'].count()
home_team_shot = filter_df_shot.groupby('homeTeam_abbrev')['type'].count()
away_team_number_of_games = df.groupby('awayTeam_abbrev')['id'].nunique()
home_team_number_of_games = df.groupby('homeTeam_abbrev')['id'].nunique()
teams = away_team_shot.index
d = {}
for team in teams:
    d[team] = (away_team_shot[team] + home_team_shot[team]) / (away_team_number_of_games[team] + home_team_number_of_games[team])

print(d)
```
